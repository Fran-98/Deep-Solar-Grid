```python
import os
import shutil
```

```python
# Ruta a la carpeta que contiene las carpetas de datos hist√≥ricos
path = 'datos' 

# Diccionario para mapear el n√∫mero del mes a su nombre en espa√±ol
meses = {
    "01": "Enero", "02": "Febrero", "03": "Marzo", "04": "Abril",
    "05": "Mayo", "06": "Junio", "07": "Julio", "08": "Agosto",
    "09": "Septiembre", "10": "Octubre", "11": "Noviembre", "12": "Diciembre"
}

# --- Inicio del Script ---
print("Iniciando la organizaci√≥n de archivos .xlsx...")

try:
    carpetas_de_origen = os.listdir(path)
except FileNotFoundError:
    print(f"‚ùå Error: La carpeta '{path}' no existe. Por favor, cr√©ala o corrige la ruta.")
    exit()

for carpeta_origen_nombre in carpetas_de_origen:
    ruta_carpeta_origen = os.path.join(path, carpeta_origen_nombre)

    # Ignorar si no es una carpeta con el formato esperado
    if not os.path.isdir(ruta_carpeta_origen) or not carpeta_origen_nombre.startswith('Datos_hist√≥ricos_'):
        continue

    try:
        # 1. Determinar la carpeta de destino (ej: datos/2024_Enero)
        fecha_str = carpeta_origen_nombre.split('_')[2]
        a√±o = fecha_str[0:4]
        mes_num = fecha_str[4:6]
        nombre_carpeta_destino = os.path.join(path, f"{a√±o}_{meses[mes_num]}")

        # 2. Crear la carpeta de destino si no existe
        os.makedirs(nombre_carpeta_destino, exist_ok=True)
        
        # 3. Determinar la ruta real de los archivos (manejando la subcarpeta extra)
        ruta_real_de_archivos = ruta_carpeta_origen
        contenido_origen = os.listdir(ruta_carpeta_origen)

        # Si hay 1 solo elemento Y ESE ELEMENTO ES UNA CARPETA, esa es la nueva ruta
        if len(contenido_origen) == 1:
            ruta_potencial_subcarpeta = os.path.join(ruta_carpeta_origen, contenido_origen[0])
            if os.path.isdir(ruta_potencial_subcarpeta):
                print(f"  -> Detectada subcarpeta anidada en '{carpeta_origen_nombre}'. Entrando...")
                ruta_real_de_archivos = ruta_potencial_subcarpeta

        # 4. Mover √∫nicamente los archivos .xlsx desde la ruta correcta
        print(f"üìÇ Buscando archivos .xlsx en: '{os.path.basename(ruta_real_de_archivos)}'")
        for archivo_individual in os.listdir(ruta_real_de_archivos):
            # Mover solo si el archivo termina en .xlsx (insensible a may√∫sculas/min√∫sculas)
            if archivo_individual.lower().endswith('.xlsx'):
                ruta_origen_archivo = os.path.join(ruta_real_de_archivos, archivo_individual)
                ruta_destino_archivo = os.path.join(nombre_carpeta_destino, archivo_individual)
                
                shutil.move(ruta_origen_archivo, ruta_destino_archivo)
                print(f"    -> ‚úÖ Movido: {archivo_individual}")

    except Exception as e:
        print(f"‚ö†Ô∏è Error procesando la carpeta '{carpeta_origen_nombre}': {e}")

print("\n¬°Proceso completado! Archivos .xlsx organizados. üéâ")
```
