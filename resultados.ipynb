```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'lstm_simple_no_met': {'RMSE': 8.2547, 'MAE': 4.6813},
    'cnn_lstm_no_met': {'RMSE': 8.3374, 'MAE': 4.6002},
    'seq2seq_lstm_no_met': {'RMSE': 7.4612, 'MAE': 3.9858},
    'lstm_simple_met': {'RMSE': 8.1814, 'MAE': 4.6719},
    'cnn_lstm_met': {'RMSE': 7.8088, 'MAE': 4.4628},
    'seq2seq_lstm_met': {'RMSE': 6.8447, 'MAE': 3.6104},
}

model_names = list(data.keys())
rmse_values = [data[name]['RMSE'] for name in model_names]
mae_values = [data[name]['MAE'] for name in model_names]

labels = [
    'LSTM (Sin Datos Meteo)',
    'CNN-LSTM (Sin Datos Meteo)',
    'Seq2Seq (Sin Datos Meteo)',
    'LSTM (Con Datos Meteo)',
    'CNN-LSTM (Con Datos Meteo)',
    'Seq2Seq (Con Datos Meteo)',
]

x = np.arange(len(labels))  # La posicin de las barras
width = 0.35  # Ancho de las barras

# --- Grfico de Barras para RMSE ---
plt.figure(figsize=(12, 7))
plt.bar(x, rmse_values, color=['#2a52be', '#2a52be', '#2a52be', '#a7c7e7', '#a7c7e7', '#a7c7e7'])
plt.ylabel('RMSE', fontsize=12)
plt.title('Comparacin de RMSE en el Conjunto de Prueba', fontsize=14, fontweight='bold')
plt.xticks(x, labels, rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Aadir los valores de las barras en el grfico para mayor claridad
for i, v in enumerate(rmse_values):
    plt.text(i, v + 0.1, f'{v:.2f}', ha='center', fontweight='bold', fontsize=10)

plt.show()

# --- Grfico de Barras para MAE ---
plt.figure(figsize=(12, 7))
plt.bar(x, mae_values, color=['#2a52be', '#2a52be', '#2a52be', '#a7c7e7', '#a7c7e7', '#a7c7e7'])
plt.ylabel('MAE', fontsize=12)
plt.title('Comparacin de MAE en el Conjunto de Prueba', fontsize=14, fontweight='bold')
plt.xticks(x, labels, rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Aadir los valores de las barras en el grfico para mayor claridad
for i, v in enumerate(mae_values):
    plt.text(i, v + 0.05, f'{v:.2f}', ha='center', fontweight='bold', fontsize=10)

plt.show()
```
